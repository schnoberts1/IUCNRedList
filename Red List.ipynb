```python
threats={'1': 'blah'}

def get_txt(key, definition):
    fq = [definition]
    last_dot = key.rfind('.')
    if last_dot == -1: 
        return fq
    parent_key = key[:last_dot]
    fq.extend(threats[parent_key])
    return fq

threats['1'] = get_txt('1', 'aaa')
threats['1.1'] = get_txt('1.1', 'bbb')
threats['1.2'] = get_txt('1.1', 'ddd')
threats['1.1.1'] = get_txt('1.1.1', 'ccc')

print(threats)


    
```

```python
threats={}

for t in threat_classification_scheme.split('\n'):
    t = t.strip()
    if len(t):
        code = t[:t.find(' ')]
        txt = t[t.find(' ')+1:]
        first_dot = code.find('.')
        threats[code] = get_txt(code, txt)
```

```python
extinct = species[(species.category == 'EX') | (species.category == 'EW')]
```

```python
print (extinct.head())
```

```python
enriched_extinct = []
to_go = len(extinct)
for index, row in extinct.iterrows():
    if to_go % 10 == 0:
        print(f'Fetching data for {row.taxonid}, {to_go} to go')
    r = requests.get(f'{RED_LIST_URL}/species/id/{row.taxonid}?token={TOKEN}')
    info = r.json()['result'][0]
    info['category'] = row.category
    info['population'] = row.population
    r = requests.get(f'{RED_LIST_URL}/species/history/id/{row.taxonid}?token={TOKEN}')
    info['assessments'] = r.json()['result'][0]
    r = requests.get(f'{RED_LIST_URL}/threats/species/id/{row.taxonid}?token={TOKEN}')
    if len(r.json()['result']):
        info['threats'] = r.json()['result'][0]
    else:
        info['threats'] = {}
    r = requests.get(f'{RED_LIST_URL}/species/narrative/id/{row.taxonid}?token={TOKEN}')
    if len(r.json()['result']):
        info['taxonomicnotes'] = r.json()['result'][0]['taxonomicnotes']
        info['populationtrend'] = r.json()['result'][0]['populationtrend']
        info['narrative_threats'] = r.json()['result'][0]['threats']
        info['rationale'] = r.json()['result'][0]['rationale']
    else:
        info['taxonomicnotes'] = None
        info['populationtrend'] = None
        info['narrative_threats'] = None
        info['rationale'] = None
        
    enriched_extinct.append(info)
    to_go -= 1
    
enriched_extinct = pd.DataFrame(enriched_extinct)
```
